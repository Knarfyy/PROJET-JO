# Création des fonctions pour analyses de tous les fichiers

```python
def traitement_fichier_licences(lien):

    # Je lis le CSV en sélectionnant seulement les colonnes pertinentes
    df = pd.read_csv(lien, sep=';', usecols=["Fédération", "Année", "Sexe", "TrancheAge", "GrandeTrancheAge", "Région", "Département", "Licences annuelles"])

    # Je filtre les sports olympiques grâce à la liste fournie par le gouvernement
    df["Olympique"] = df["Fédération"].apply(lambda x: "Yes" if any(sport in x.lower() for sport in liste_fede_olympique) else "No")
    df_olympique = df[df["Olympique"] == "Yes"]

    # Je supprime les Fédé non filtrées comme prévues
    df_olympique_2 = df_olympique[~df_olympique["Fédération"].isin([
        "Fédération Française d'Éducation Physique et de Gymnastique Volontaire", 
        "Fédération Française de Motocyclisme",
        "Fédération Française de Savate, Boxe Française et Disciplines Associées",
        "Fédération Française de Ski Nautique et de Wakeboard",
        "Fédération Française de Rugby à XIII",
        "Fédération Française de Char à Voile",
        "Fédération Française des Sports de Traîneau, de Ski VTT Joëring et de Canicross",
        "Fédération Française de Javelot Tir sur Cible"
    ])]

    # Je supprime les régions "Non réparti" & "Etranger"
    df_olympique_3 = df_olympique_2[~df_olympique_2['Région'].isin(["NR - Non réparti", "ETR - Etranger"])]

    return df_olympique_3
```

```python
def traiter_tous_fichiers_combine(dossier=".", pattern="*.csv"):
    """
    Traite tous les fichiers CSV d'un dossier et combine les résultats en un seul DataFrame
    """
    fichiers = glob.glob(os.path.join(dossier, pattern))
    
    resultats_combines = []
    
    for fichier in fichiers:
        print(f"Traitement de : {os.path.basename(fichier)}")
        try:
            df_traite = traitement_fichier_licences(fichier)
            resultats_combines.append(df_traite)
        except Exception as e:
            print(f"Erreur avec {fichier}: {e}")
    
    df_final = pd.concat(resultats_combines, ignore_index=True)
    print(f"Résultat final : {len(df_final)} lignes")

    df_final.to_csv("licences_2016-2023.csv", sep=';', index=False)
    print(f"Export réalisé")
    
    return df_final
```

```python
df_final = traiter_tous_fichiers_combine(dossier="DataSource\Licences-annuelles-par-millesime-2016-2023", pattern="*.csv")
```
